```python
#sentiment analysis

import pandas as pd
from textblob import TextBlob
file_path = "C:\\Users\\Akshat\\Downloads\\flipkart_reviews_large_dataset_sample.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')
df['reviewDescription'] = df['reviewDescription'].astype(str)
df["sentiment"] = df["reviewDescription"].apply(lambda x: TextBlob(x).sentiment.polarity)
df["sentiment_label"] = df["sentiment"].apply(lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Neutral"))
output_file_path = "C:\\Users\\Akshat\\Downloads\\updated_with_sentiments.csv"
df.to_csv(output_file_path, index=False)
print(f"Sentiment analysis results saved to {output_file_path}")
```

```python
#dvz1

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\Akshat\\Downloads\\updated_with_sentiments.csv")
plt.scatter(df['productPrice'], df['averageRating'])
plt.title("Scatter Plot: Product Price vs. Average Rating")
plt.xlabel('Product Price')
plt.ylabel('Average Rating')
plt.show()
```

```python
#dvz2

top_reviewers = df['reviewAuthor'].value_counts().nlargest(10)
top_reviewers.plot(kind='barh', color='skyblue')
plt.title("Top 10 Reviewers by Number of Reviews")
plt.xlabel('Number of Reviews')
plt.ylabel('Reviewer')
plt.show()
```

```python
#filling the missing values with mean

import pandas as pd
file_path = "C:\\Users\\Akshat\\Downloads\\flipkart_reviews_large_dataset_sample.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
cleaned_file_path = "C:\\Users\\Akshat\\Downloads\\updated_with_sentiments.csv"
df.to_csv(cleaned_file_path, index=False)
print(f"Missing values handled and saved to {cleaned_file_path}")
```

```python
#filling missing values with median

import pandas as pd
file_path = "C:\\Users\\Akshat\\Downloads\\flipkart_reviews_large_dataset_sample.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
cleaned_file_path = "C:\\Users\\Akshat\\Downloads\\updated_with_median.csv"
df.to_csv(cleaned_file_path, index=False)
print(f"Missing values handled and saved to {cleaned_file_path}")
```
